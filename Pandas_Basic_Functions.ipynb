```python
import pandas as pd
```

```python
data = {'age':[25,22,18,30,45,50,35,20,55,40],
       'gender': ['M','F', 'F', 'M', 'M', 'F', 'M','F', 'M', 'M'],
       'score':[90,85,75,95,70,85,75,90,95,85]}
df = pd.DataFrame(data)
df
```

```python
df['gender'].value_counts()
```

```python
df['gender'].value_counts(normalize=True)
```

#### 2.Where()

```python
df.where(df['age']>30)
```

```python
df.where(df['age']>30, other= 0)
```

```python
df.where(df['age']>30, other= 0).all(1)
```

```python
df[df.where(df['age']>30, other= 0).all(1)]
```

```python
f1 = df['age']>30
f2 = df['gender']=='F'
```

```python
df.where(f1&f2, other = 0)
```

```python
df[df.where(f1&f2, other = 0).all(1)].reset_index()
```
